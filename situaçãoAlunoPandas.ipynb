```python
import pandas as pd

arquivo = pd.read_csv("situaçãoAluno.txt")

media = (arquivo["nota_1"] + arquivo["nota_2"]) / 2

arquivo["media"] = media

arquivo.loc[arquivo["media"] >= 7, "situacao"] = "Aprovado"

arquivo.loc[(arquivo["media"] < 7) | (arquivo["faltas"] > 5 ), "situacao"] = "Reprovado"

arquivo["faltas"].max()

arquivo["media"].mean()

arquivo["media"].max()

arquivo.to_csv("sample_data/situacaoAlunos.txt")

arquivo
```
